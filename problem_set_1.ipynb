<b>5. </b>Write a function <code>threshold_image</code> which thresholds an image based on a threshold level given as a parameter to the function. The function should take two parameters: the image to be thresholded and the threshold level. The result of the function should be a new thresholded image. <br /> <br />

```python
#Solution

def threshold_image(src,threshold_value):
    heightth, widthth = src.shape[:2]                        #Extract image size from shape function
    imgth= src.copy()   #Copy image
    for i in range(heightth):       #Go over every row
        for j in range(widthth):    #Go over every column
            if np.any(imgth[i,j] < threshold_value):   #Find if pixel value is less than threshold value 
                imgth[i,j] = 255    #If this is the case, replace the pixel value with 255
    return imgth                    #Return img 

index = 100;

for i,img_name in enumerate(imagesnames):    #Show results for each image
    plt.figure(figsize=(15,15));
    imgfn = cv2.imread("images/"+img_name, cv2.IMREAD_COLOR)
    rth = threshold_image(imgfn,index)      #Call function
    plt.subplot(1,3,1); plt.imshow(imgfn);plt.title('Original Image');plt.axis('off') #Display the image
    plt.subplot(1,3,2); plt.imshow(rth);plt.title('Thresholded Image');plt.axis('off') #Display the image

```

<b>8. </b>Keeping your original image resolution reduce the number of gray levels in your image from 256 to 2 in powers of 2. Display your results.


```python
#Solution

def look_up_table(new_levels):   #Define function
    paso = 255.0/(new_levels - 1) #Paso entre niveles contando el 0 como primer nivel
    new_levels_mag = paso*np.arange(new_levels) #[0 1 ... new_levels-1]*paso, los niveles
    old_levels_mag = np.arange(256) #[0 1 ... 255]
    cambio = np.zeros((new_levels, 256)) #Matriz de 0 (new_levels filas, 256 columnas)
    for level in range(0,new_levels):
        cambio[level, :] = abs(new_levels_mag[level]-old_levels_mag) #Restar los niveles nuevos con los viejos
    LU_T = new_levels_mag[np.argmin(cambio,0)] #Los elementos de list_levels indicados por el vector de indices np.argmin(distance,0)
    LU_T = np.uint8(LU_T) #Convierte al formato uint8
    return LU_T

def intensity_reduce(image,new_levels):   #Define function
    imag = cv2.LUT(image, look_up_table(new_levels)) #Realiza la transformacion con los niveles de la look-up table
    return imag    #Return image

for i,img_name in enumerate(imagesnames):    #Load images
    imgfn = cv2.imread("images/"+img_name, cv2.IMREAD_COLOR)  #Read images        
    plt.figure(figsize=(15,15));   #Create figure por plots
    plt.subplot();plt.imshow(imgfn, cmap='gray'); plt.title('Original Image')   #Plot each image
    for k in range(1,8):
        img2 = intensity_reduce(imgfn, 2**(8-k))    #Apply function previously defined
        plt.suptitle(img_name, fontsize=15)  #Title every section of images
        plt.subplot(4,2,k); plt.imshow(img2, cmap='gray'); plt.title(str(256/2**k)+" Levels");plt.axis('off')  #Plot resulting image
        plt.subplots_adjust(hspace=.5)
```

<h3>- Descripción de las funciones de librerías utilizadas:</h3>
<p style="text-align:justify">
La función desarrollada en este inciso realiza los siguientes procesos: Primero se determina la distancia entre los nuevos niveles de intensidad que tendrá la imagen. Teniendo este valor, se utiliza la función "np.arange()" de la librería Numpy que toma como valor de entrada el numero de niveles de intensidad deseado para crear un vector cuyos elementos son los números desde el cero hasta el numero de niveles de intensidad menos uno. Este vector se multiplica por la distancia entre niveles para dar lugar a un vector cuyos elementos son los niveles de intensidad como tal, un vector de niveles. </p>
<p style="text-align:justify">
Con el vector de niveles ya creado, se procede a crear un vector con los niveles de intensidad originales, es decir, un vector cuyos elementos son los números del 0 al 255. Seguidamente se crea una matriz con 256 columnas, representando estas los niveles de intensidad originales, y un numero de filas igual al numero de niveles deseado, cada fila representando un nivel de intensidad. Las 256 columnas de cada fila se llenan con el valor absoluto de la diferencia entre el nivel correspondiente a la fila y cada uno de los 256 niveles originales, de modo que cada celda de la matriz indica que tan cercano esta el nuevo nivel indicado por la fila al nivel original indicado por la columna.</p>
<p style="text-align:justify">
A la matriz de diferencias se le aplica la función "np.argmin()" de la librería Numpy que arroja el valor mínimo de un vector o una matriz. Esta función recibe como parámetros la matriz o vector y en caso de ser una matriz, se puede definir un eje según el cual se encontraran los mínimos, si se encontraran por filas o por columnas. Si el eje es igual a 0, como en este caso, la función devuelve un vector con el indice de los valores mínimos por columna, la fila en la que se encuentra el valor, efectivamente indicando a que nivel nuevo se acerca mas el nivel viejo que indica la columna. </p>
<p style="text-align:justify">
Con el vector de indices y el vector de niveles obtenidos, es posible crear una look-up table. Esto se logra creando un nuevo vector que contenga los valores del vector de niveles indicados por las posiciones en el vector de indices, es decir, un vector de 256 posiciones que indica que nivel de intensidad nuevo corresponde al nivel de intensidad original, siendo el indice el nivel original y el elemento guardado en esa posición, el nivel nuevo.  Finalmente se llama a la función CV2.LUT() de la librería OPENCV, que recibe como parámetros una imagen y una look-up table. Esta función toma los valores de cada pixel de la imagen y los reemplaza por el valor correspondiente en la look-up table por lo que de esta forma se obtiene la imagen con los niveles de intensidad reducidos deseados. </p>
